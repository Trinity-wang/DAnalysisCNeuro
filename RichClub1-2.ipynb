```python
# For this analysis, the Hybrid Connectivity Matrix was turned in 2 csv files. The 2s were converted to ones, 
#and turned into a csv file and the -1 and -2s were 
# added to separate matrix about excitatory and inhibitory function. For this analysis, the first matrix was used. 
# Then, the data was sorted, and each type of data (example - CA1 inputs, or DG outputs) turned into it's own Excel file.
# The columns of each entry were then summed. Then the intermediate data was deleted so only an index number and the 
# number of connections remained.
# AFter this, the below code was created to make scatter plots so the output and input connections 
# could be compared for each section
# The "sub" data was not used, because no input cells were found and only 2 output cells were found.

# input connections are in orange
# output connectinos are in blue

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd # these lines denote the required templates to create the scatter plots
data1 = pd.ExcelFile("CA1+.xlsx").parse() # loads the data
data2 = pd.ExcelFile("CA1-.xlsx").parse() # loads the data

x1 = data1 # gives data a variable name
x2 = data2 # gives data a variable name


plt.scatter(x1.columns, x1,marker='v') # plots the outputs as an upward pointing triangle
plt.scatter(x2.columns, x2,marker='^') # plots the inputs as a downward pointing triangle
plt.title('CA1') # adds a title
plt.xlabel('x') # adds an x axis label
plt.ylabel('Number of Connections') # adds a y axis label
plt.show() # makes the plot show up




print(data1) #prints the data 
print(data2)
```
