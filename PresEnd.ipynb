```python
import numpy as np
import os

import glob
toggle_files = glob.glob('buffer/*raster*.png')
lenof = len(toggle_files)
def plot_raster(tf):
    import glob
    #from natsort import natsorted
    toggle_files = glob.glob('buffer/*raster*.png')
    toggle_files = sorted(toggle_files)
    #print(toggle_files)
    keyed = { k:v for k,v in enumerate(toggle_files) }

    display(Image(str(keyed[tf])))

    
interact(plot_raster, tf = widgets.IntSlider(
    value=0,
    min=0,
    max=lenof-1,
    step=1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
))
```
